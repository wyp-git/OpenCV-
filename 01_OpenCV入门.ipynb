```python
import cv2
import matplotlib.pyplot as plt
```

## 1.读取和显示图片

```python
img = cv2.imread("img/image.jpg", 1)
bgr_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(bgr_img)
plt.show()
```

## 2.保存图片

```python
cv2.imwrite("img/img_write.jpg", img)
```

## 3.图像压缩
- jpg有损压缩，无透明度，压缩范围0~100
- 0：压缩效果最差

```python
cv2.imwrite("img/img_quality.jpg", img, [cv2.IMWRITE_JPEG_QUALITY, 50])
bgr_quality_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(bgr_quality_img)
plt.show()
```

- png无损压缩,有透明度，压缩范围0~9
- 0：压缩效果最好

```python
cv2.imwrite("img/img_png_0.png", img, [cv2.IMWRITE_PNG_COMPRESSION, 0])
bgr_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(bgr_img)
plt.show()
```

```python
cv2.imwrite("img/img_png_9.png", img, [cv2.IMWRITE_PNG_COMPRESSION, 9])
```

## 3.操作像素

```python
(b, g, r) = img[100, 100]
print(b, g, r)
```

```python
for i in range(100):
    img[10+i, 100] = (0, 0, 0)
bgr_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(bgr_img)
plt.show()
```
