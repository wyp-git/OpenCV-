```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imge.jpg",1)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
b, g, r = cv2.split(img)
b, g, r = 255- b, 255 - g, 255 - r
neg_img = np.zeros((img.shape[0], img.shape[1], 3), np.uint8)
neg_img[:,:,0] = b
neg_img[:,:,1] = g
neg_img[:,:,2] = r

neg_rgb_img = cv2.cvtColor(neg_img, cv2.COLOR_BGR2RGB)
plt.imshow(neg_rgb_img)
plt.show()
```
