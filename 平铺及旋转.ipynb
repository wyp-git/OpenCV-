```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imge.jpg", 1)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

- 

```python
w = img.shape[1]
h = img.shape[0]

#5
sz1 = w * 5 
#2
sz0 = h * 2

#
dst = np.zeros((sz0, sz1, 3), np.uint8)

#
img_x = 0
img_y = 0
for n_y in range(0, sz0):
    for n_x in range(0, sz1):
        dst[n_y, n_x, 0] = img[img_y, img_x, 0]
        dst[n_y, n_x, 1] = img[img_y, img_x, 1]
        dst[n_y, n_x, 2] = img[img_y, img_x, 2]
        
        #
        img_x += 1
        if img_x >= w:
            img_x = 0
            
    #
    img_y += 1
    if img_y >= h:
        img_y = 0
        
rgb_dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_dst)
plt.show()
```

```python
img = cv2.imread("imge.jpg", 1)
w = img.shape[1]
h = img.shape[0]
#2
sz1 = w * 2 
#5
sz0 = h * 5

dst = np.zeros((sz1, sz0, 3), np.uint8)

#
img_x = 0
img_y = 0
for n_y in range(0, sz0):
    for n_x in range(0, sz1):
        dst[n_x, n_y, :] = img[img_y, img_x, :]
        
        #
        img_x += 1
        if img_x >= w:
            img_x = 0
            
    #
    img_y += 1
    if img_y >= h:
        img_y = 0
        
rgb_dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_dst)
plt.show()
```
