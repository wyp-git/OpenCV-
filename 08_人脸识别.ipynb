```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### 加载XML文件

```python
face_xml = cv2.CascadeClassifier("resources/haarcascade_frontalface_default.xml")
eye_xml = cv2.CascadeClassifier("resources/haarcascade_eye.xml")
```

## 加载图片

```python
img = cv2.imread("img/face.jpg", 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot_gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
plt.figure()
plt.imshow(plot_img)
plt.figure()
plt.imshow(plot_gray)
plt.show()
```

- 人脸识别

```python
faces = face_xml.detectMultiScale(gray, 1.3, 5)
len(faces)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2) #脸
    roi_face = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    
    eyes = eye_xml.detectMultiScale(roi_face) 
    for (e_x, e_y, e_w, e_h) in eyes:
        cv2.rectangle(roi_color, (e_x, e_y), (e_x+e_w, e_y+e_h), (0,255, 0), 2) #眼睛

plot_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(plot_img)
plt.show()
```
